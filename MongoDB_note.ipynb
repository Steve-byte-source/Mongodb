## Use DB

```python
db_Stock = client.Stock
print(db_Stock)
```

## Prepare JSON to Insert Into MongoDB

```python
stock_df = pd.read_csv("C:/Users/User/Desktop/dealer_information.csv", index_col=0, encoding="utf-8")

insert_list = list()
key_list = list()
for key in stock_df:
    key_list.append(key)


for row_num in range(len(stock_df)):
    stock_dict = dict()
    for index in range(len(key_list)):
        if key_list[index] == 'location':
            lat = stock_df.iloc[row_num][index].split("[")[1].split("]")[0].split(",")[0]
            lon = stock_df.iloc[row_num][index].split("[")[1].split("]")[0].split(",")[1]            
            location = [float(lat), float(lon)]
            # print(location)
            stock_dict[key_list[index]] = {"type": "Point", "coordinates": location}
        else:
            stock_dict[key_list[index]] = stock_df.iloc[row_num][index].strip("")
    
    print(stock_dict)
    insert_list.append(stock_dict)
    # 匯入List後清除
    stock_dict = dict()
```
